### Low Big_BETS DDR Mutations

```python
#low bigbets DDR genes survival analysis
outfile=os.path.join(graphics_dir,'samstein_survival_low_zscore.pdf')

cph = lifelines.CoxPHFitter()
# vars2keep=['high_TMB','DDR','DDR:highTMB']
vars2keep=['DDR_low_z']

dummy_df=samstein.clinical_bigbets_df.loc[:,['os','censOS']]

for var in vars2keep:
    cdf=pd.get_dummies(samstein.clinical_bigbets_df.loc[:,var])
    cdf.columns=map(lambda x: var+"_"+str(x),cdf.columns.values)
    dummy_df=pd.concat([dummy_df,cdf.iloc[:,:-1]],axis=1)

dummy_df['log(TMB)']=np.log1p(samstein.clinical_bigbets_df.loc[:,'TMB'])
# dummy_df['TMB']=samstein.clinical_bigbets_df.loc[:,'high_TMB']

dummy_df['DDRlow:highTMB']=samstein.clinical_bigbets_df.loc[:,'DDRlowz:highTMB']

res=cph.fit(dummy_df, 'os', event_col='censOS')
llrstat=cph.log_likelihood_ratio_test()
llr=llrstat.test_statistic
df=llrstat.degrees_freedom
p_value=llrstat.p_value
summary=res.summary
print(res.summary)

var='DDR_low_z_TMB'



coms=samstein.clinical_bigbets_df[var].unique()
commap=dict(zip(coms,range(len(coms))))
clust=list(map(lambda x: commap[x],samstein.clinical_bigbets_df[var]))

plt.close()
f,a=plt.subplots(1,2,figsize=(7.5,10))

nrows=10
ncols=1

ncols1=1
nrows1=6

a=plt.subplot2grid((nrows,ncols),(0,0),rowspan=nrows1,colspan=1)

cols = [red + (1.0,), red + (.5,), blue + (1.0,), blue + (.5,)]
cols=[red+(1.0,),red+(.7,),blue+(1.0,),blue+(.7,)]

kmf = lifelines.KaplanMeierFitter()
df2test=samstein.clinical_bigbets_df
i=0
for name,gdf in df2test.groupby(var):
    if re.search("WT",name):
        ls=':'
    else:
        ls='-'
    kmf.fit(gdf["os"], gdf["censOS"], label=name)
    kmf.plot(ax=a,ci_show=False,color=cols[i],linestyle=ls,
             show_censors=True, censor_styles={'ms': 5, 'marker': '+'})
    i+=1
# lg_lines=lg.get_lines()
# for j,ln in enumerate(a.get_lines()):
#     ln.set_color(cols[j])
#     lg_lines[j].set_color(cols[j])

handles, labels = a.get_legend_handles_labels()
new_labels=['Low-BiG-BETS DDR Mut','Low-BiG-BETS DDR WT','Low-BiG-BETS DDR Mut','Low-BiG-BETS DDR WT']
l=a.legend(loc=1,bbox_to_anchor=(.85,1.0),fontsize=14,
           frameon=False,handles=handles,labels=new_labels)
for i,text in enumerate(l.get_texts()):
    plt.setp(text, color = cols[i])
    
a.annotate(s='TMB-H',fontsize=14,xy=(.83,.925),xytext=(.86,.925),
       arrowprops=dict(arrowstyle='-[, widthB=1.1, lengthB=0', lw=1.0),
       xycoords='axes fraction',verticalalignment='center',transform=a.transAxes)

a.annotate(s='TMB-L',fontsize=14,xy=(.83,.825),xytext=(.86,.825),
       arrowprops=dict(arrowstyle='-[, widthB=1.1, lengthB=0', lw=1.0),
       xycoords='axes fraction',verticalalignment='center',transform=a.transAxes)

a.spines['right'].set_visible(False)
a.spines['top'].set_visible(False)

a.set_xlabel("months",fontsize=20)
a.set_ylabel("overall survival",fontsize=20)

# a.set_title("Kaplan-Meier Curve by TMB + low TMB Assoc.DDR Genes",fontsize=18,fontweight='bold')
pval_df=bigbets.bipartite_helper_functions.get_os_stats(clust=clust,samp_df=samstein.clinical_bigbets_df)
pval=pval_df['llr_pval'].values[0]
a.text(x=.85,y=.05,s="CPH pval={:.1e}".format(pval),horizontalalignment='center',
       fontsize=14,transform=a.transAxes)



a=plt.subplot2grid((nrows,ncols),(nrows1,0),rowspan=nrows-nrows1,colspan=1)
# a=plt.subplot(1,2,2)
# a.text(s='B)',fontsize=18,fontweight='bold',x=0.0,y=.95,transform=a.transAxes)
# a.text(s='B)',fontsize=18,fontweight='bold',x=-.07,y=1.0,transform=a.transAxes)

xlim_min=np.min(summary['coef']-1.96*summary['se(coef)'])
xlim_max=np.max(summary['coef']+1.96*summary['se(coef)'])

for i,ind in enumerate(summary.index):
    expcoef=summary.loc[ind,'exp(coef)']
    up=summary.loc[ind,'exp(coef) upper 95%']-expcoef
    low=expcoef-summary.loc[ind,'exp(coef) lower 95%']
    ebars=np.array([[low,up]]).T
    a.errorbar(x=[expcoef],alpha=.7,
               y=[summary.shape[0]-1-i],marker='s',color='k',uplims=True, lolims=True,
               xerr=ebars
#                xerr=[1.96*summary.loc[ind,'se(coef)'],
                    )

    
a.vlines(x=1.0,linestyle='--',alpha=.6,linewidth=1,
         ymin=0,ymax=summary.shape[0]-.9)

a.set_xscale('log')
a.set_xlim([.0000025,30])
a.spines['bottom'].set_position(('data', -.4))
a.spines['bottom'].set_bounds( 0.1,5)

a.spines['left'].set_visible(False)
a.spines['right'].set_visible(False)
a.spines['top'].set_visible(False)
a.set_yticks([])
a.set_xticks([.1,1,5])
a.xaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
plt.minorticks_off()
a.set_ylim([-2,3])


# a.text(s="LLR pval={:.2e}".format(p_value),fontsize=18,
#        x=.2,y=.8,transform=a.transAxes)



axis_to_data = a.transAxes + a.transData.inverted()
data_to_axis = axis_to_data.inverted()

y=data_to_axis.transform((.1,summary.shape[0]-.55))[1]

print('y',y)
a.text(s="Covariate".format(p_value),fontsize=20,
       x=.01,y=y,transform=a.transAxes)

a.text(s="HR (95% CI)".format(p_value),fontsize=20,
       x=.375,y=y,transform=a.transAxes)

a.text(s="P-value".format(p_value),fontsize=20,
       x=.85,y=y,transform=a.transAxes)

a.hlines(y=y-.02,xmin=-2,xmax=50,transform=a.transAxes,linewidth=3,linestyle='-')
# a.hlines(y=y-.03,xmin=-1,xmax=2,transform=a.transAxes,linewidth=1,linestyle='-')

for i,ind in enumerate(summary.index):
    y=data_to_axis.transform((.1,summary.shape[0]-i-1.2))[1]
    print('y',y)
    if summary.loc[ind,'p']<.05:
        a.text(s="{}*".format(ind),fontsize=20,
       x=-.1,y=y,transform=a.transAxes)
    else:
        a.text(s=ind,fontsize=20,
           x=-.1,y=y,
           transform=a.transAxes)
    hrtext="{:.2f}  ({:.2f}-{:.2f})".format(*summary.loc[ind,['exp(coef)',
                                                         'exp(coef) lower 95%','exp(coef) upper 95%']])
    pvaltext="{:.2e}".format(*summary.loc[ind,['p']])
    a.text(s=hrtext,fontsize=20,
       x=.35,y=y,
           transform=a.transAxes)
    a.text(s=pvaltext,fontsize=20,
       x=.85,y=y,
           transform=a.transAxes)
    
    x1,y1=data_to_axis.transform((.01,summary.shape[0]-i*.9-1.5))
    x2,y2=data_to_axis.transform((.01,summary.shape[0]-i*.9-.8))
    

    if i%2==0:
        crect=patches.Rectangle((-.1,y1),width=1.2,height=y2-y1,transform=a.transAxes,fill=True,
                               alpha=.05,color='k',clip_on=False)
#         crect2=patches.Rectangle((x2,y1),width=1.1-x2,height=y2-y1,transform=a.transAxes,fill=True,
#                                alpha=.05,color='k')
        a.add_patch(crect)
#         a.add_patch(crect2)
        
    
    

# a.set_title("CPH Model Survival",fontsize=18,fontweight='bold')
# cph.plot(ax=a)
a.text(s="Hazard Ratio",fontsize=18,x=1.25,
       y=-1.1,
       horizontalalignment='center')

f.suptitle('Samstein',fontsize=22,x=.5,y=.97,transform=f.transFigure)

f.subplots_adjust(top=.93,bottom=-.05,right=.95,left=.101,wspace=.0,hspace=1.5)
f.savefig(outfile,dpi=200)
plt.show()



```
